# Campo Electrico

Armamos una clase que devuel el campo electrico de un carga

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

dim= 128
ny= 128
nx= 128

def E(q, r0, x, y):

    """Retorna el vector de campo eléctrico E=(Ex,Ey) de una carga q en r0"""

    den = np.hypot(x-r0[0], y-r0[1])**3

    return q * (x - r0[0]) / den, q * (y - r0[1]) / den

q=1
cargas = []
cargas.append((q, (-10,0)))
cargas.append((-q, (10,0)))


vx, vy = np.zeros((ny, nx)), np.zeros((ny, nx))

x = np.linspace(-40, 40, nx)

y = np.linspace(-40, 40, ny)

X, Y = np.meshgrid(x, y)

for carga in cargas:

    ex, ey = E(*carga, x=X, y=Y)

    vx += ex

    vy += ey


```

```python
# plot
plt.figure()
plt.streamplot(x,y, vx, vy)
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.grid(True)
plt.show()
```
