```python
import numpy as np
import matplotlib.pyplot as plt
```

# Dipolos largos

<img src=".\1.png">

```python
tetha=np.linspace(0,np.pi,1000,endpoint=False)
factores=[0.5,1,1.5,5]

plt.figure(dpi=150)
for factor in factores:

    F=(np.cos(np.pi*factor*np.cos(tetha))-np.cos(np.pi*factor))**2/(np.sin(tetha)**2)
    plt.plot(tetha,F,label="L/Lambda= {}".format(factor))
    plt.legend(loc='upper right')
plt.grid(True)
plt.title("Factor de configuraci√≥n")


```

```python
tetha=np.linspace(0,2*np.pi,1000,endpoint=False)
factores=[0.5,1,1.5,5]
for factor in factores:
    plt.figure(dpi=100)
    F=((np.cos(np.pi*factor*np.cos(tetha))-np.cos(np.pi*factor))**2/(np.sin(tetha)**2))
    plt.polar(tetha,F,label="L/Lambda= {}".format(factor))
    plt.legend(loc='upper right')
    plt.title("Factor de configuraci√≥n")
```

# Antena de onda progresiva

<img src=".\2.png">
<img src=".\3.png">

```python
tetha=np.linspace(0,2*np.pi,1000,endpoint=False)
factores=[0.5,1,1.5,5]
plt.figure(dpi=100)
for factor in factores:

    F=(np.sin(np.pi*factor*(1-np.cos(tetha)))**2)/(np.tan(tetha/2)**2)
    plt.polar(tetha,F,label="L/Lambda= {}".format(factor))
    plt.legend(loc='upper right')
    plt.title("Factor de configuraci√≥n")

for factor in factores:
    plt.figure(dpi=100)
    F=(np.sin(np.pi*factor*(1-np.cos(tetha)))**2)/(np.tan(tetha/2)**2)
    plt.polar(tetha,F,label="L/Lambda= {}".format(factor))
    plt.legend(loc='upper right')
    plt.title("Factor de configuraci√≥n")
```

```python
phi=np.linspace(0,2*np.pi,1000,endpoint=False)

#### Constantes ####
c=3e8
fcia=1e6
Lambda=c/fcia
beta=2*np.pi/Lambda

d=Lambda/4
xi=-np.pi/2

A=np.abs(np.cos(0.5*(beta*d*np.cos(phi)+xi)))
#A=np.abs(np.cos(np.pi/4*np.cos(phi)))

#psi=beta*d*np.cos(phi)+xi


print("d",d)
print("xi",xi)
print("beta . d",beta*d)
```

```python
from ipywidgets import interact, interactive, fixed, interact_manual
```

```python
def f(d_Lambda, xi):
    plt.figure()
    plt.polar(phi,np.abs(np.cos(0.5*(2*np.pi*d_Lambda*np.cos(phi)+xi))))
    plt.show()

interactive_plot = interactive(f, d_Lambda=(0,1,0.05), xi=(-np.pi/2,np.pi/2, np.pi/4))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
