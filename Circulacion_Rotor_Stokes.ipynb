# Circulacion e integral de linea

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

# set up a normalized grid:
dim= 10
xarray= np.arange(-dim,dim)
yarray= np.arange(-dim,dim)

x,y = np.meshgrid(xarray,yarray)
```

# Trabajo hecho por la fuerza peso

El campo de fuerza de gravedad como un vector de fuerza que vale (x,y)=(0,-1) en todo los puntos del espacio

```python
vx = (y*0)
vy = (x*0)-1

print(np.shape(vx),np.shape(vy),np.shape(x),np.shape(y))
print((vx[0][0],vy[0][0]))
```

```python
# plot the flow lines:
plt.figure()
plt.streamplot(x,y, vx, vy)
plt.xlabel("[m]")
plt.ylabel("[m]")
plt.axis('equal')
plt.title("Fuerza [N]")
plt.show()
```

para calcular el trabajo al rededor de una curva cerrada definimos el contorno

```python
xmin=5
xmax=15
ymin=5
ymax=15
```

la integral de linea cerrada se reduce a una suma

```python
# La circulacion

bottom = -np.sum(vy[ymin,xmin:xmax]*(xmax-xmin))
print("la circulacion por la base",bottom)
right = np.sum(vx[ymin:ymax,xmax]*(ymax-ymin))
print("la circulacion por la derecha",right)
top = np.sum(vy[ymax,xmax:xmin:-1])*(xmax-xmin)
print("la circulacion por la tapa", top,)
left = -np.sum(vx[ymax:ymin:-1,xmin])*(ymax-ymin)
print("la circulacion por la izquierda",left,)

print("suma de las circulaciones",(top+bottom+left+right))
circulacion = (top+bottom+left+right)
```

Cuando el camino perpendicular a la fuerza no hace trabajo y cuando va en contra o a favor, hay trabajo

Tracemos el flujo y el cuadro, donde evaluamos el trabajo:

```python
plt.figure()
plt.title('La integral de linea por el camino cerrado = {} Joules'.format(circulacion))

plt.streamplot(x,y, vx, vy)

# la cajita
plt.plot(x[ymin,xmin:xmax+1],y[ymin,xmin:xmax+1],'r',alpha=0.5) # bottom
plt.plot(x[ymax,xmax:xmin-1:-1],y[ymax,xmax:xmin-1:-1],'r',alpha=0.5) # top
plt.plot(x[ymax:ymin-1:-1,xmin],y[ymax:ymin-1:-1,xmin],'r',alpha=0.5) # left
plt.plot(x[ymin:ymax+1,xmax],y[ymin:ymax+1,xmax],'r',alpha=0.5) # right
plt.xlabel("eje $x$")
plt.ylabel("eje $y$")
plt.axis('equal')
plt.show()
```

# Rotor de un campo vectorial

Quizás el ejemplo más intuitivo sea con fluidos. Este flujo representa un campo vectorial, en el que podría insertar una rueda con paletitas imaginarias para describir la presencia de remolinos o rotación en el campo. Estas ruedas de paletas representan bien la interpretación física del operador de rotor:

"El componente rotor de un campo vectorial ${\bf v}$ en una determinada dirección es la integral de línea cerrada de ${\bf v}$ a lo largo de una trayectoria cerrada perpendicular a esta dirección, por unidad de superficie". Esto es



$$ \frac{\oint {\bf v} \cdot d{\bf S}}{d{\bf S}}$$

Establezcamos una cuadrícula para el flujo de fluidos, donde el rotor de este campo vectorial en particular también se llama vorticidad
(https://en.wikipedia.org/wiki/Vorticity):

```python
vx = -y
vy = x
```

```python
# plot the flow lines:
plt.figure()
plt.streamplot(x,y, vx, vy)
plt.xlabel("[m]")
plt.ylabel("[m]")
plt.axis('equal')
plt.title("Fuerza [N]")
plt.show()
```

Para definir el rotor hacemos tender la superfie a cero... eso en calculo... aca al valor de area mas chico

```python
xmin=5
xmax=6
ymin=5
ymax=6
```

la integral de linea cerrada se reduce a una suma

```python
bottom = -np.sum(vy[ymin,xmin:xmax])
print("la circulacion por la base",bottom,"Joules")
right = np.sum(vx[ymin:ymax,xmax])
print("la circulacion por la derecha",right,"Joules")
top = np.sum(vy[ymax,xmax:xmin:-1])
print("la circulacion por la tapa", top,"Joules")
left = -np.sum(vx[ymax:ymin:-1,xmin])
print("la circulacion por la izquierda",left,"Joules")
area=((xmax-xmin)*(ymax-ymin))

print("suma de las circulaciones",(top+bottom+left+right))
rotor = (top+bottom+left+right)/area

print("el calculo del rotor da",rotor)
```

```python
plt.figure()
plt.title('calculo del rotor en el cuadrado rojo = {} '.format(rotor))

plt.streamplot(x,y, vx, vy)

# la cajita
plt.plot(x[ymin,xmin:xmax+1],y[ymin,xmin:xmax+1],'r',alpha=0.5) # bottom
plt.plot(x[ymax,xmax:xmin-1:-1],y[ymax,xmax:xmin-1:-1],'r',alpha=0.5) # top
plt.plot(x[ymax:ymin-1:-1,xmin],y[ymax:ymin-1:-1,xmin],'r',alpha=0.5) # left
plt.plot(x[ymin:ymax+1,xmax],y[ymin:ymax+1,xmax],'r',alpha=0.5) # right
plt.xlabel("eje $x$")
plt.ylabel("eje $y$")
plt.axis('equal')
plt.show()
```

el rotor de este campo vectorial es igual a dos para todo el plano XY. Se puede verificar analiticamente y variando la posicion de la caja.


# Teorema de Stokes

Sea S una superficie orientada por un vector unitario normal n, acotada por una curva cerrada simple suave a trozos C con orientación positiva. Si F es un campo vectorial con derivadas parciales continuas en una región abierta que contiene a S y a C entonces


$${\displaystyle \oint _{\partial S}\mathbf {F} \cdot d\mathbf {r} =\iint _{S}\left(\nabla \times \mathbf {F} \right)\cdot d\mathbf {S} }$$



```python
xmin=4
xmax=16
ymin=4
ymax=16
```

la integral de linea cerrada se reduce a una suma

```python
# La circulacion

bottom = -np.sum(vy[ymin,xmin:xmax]*(xmax-xmin))
print("la circulacion por la base",bottom)
right = np.sum(vx[ymin:ymax,xmax]*(ymax-ymin))
print("la circulacion por la derecha",right)
top = np.sum(vy[ymax,xmax:xmin:-1])*(xmax-xmin)
print("la circulacion por la tapa", top,)
left = -np.sum(vx[ymax:ymin:-1,xmin])*(ymax-ymin)
print("la circulacion por la izquierda",left,)

print("suma de las circulaciones",(top+bottom+left+right))
circulacion = (top+bottom+left+right)
```

```python
plt.figure()
plt.title('La integral de linea por el camino cerrado = {} '.format(circulacion))

plt.streamplot(x,y, vx, vy)

# la cajita
plt.plot(x[ymin,xmin:xmax+1],y[ymin,xmin:xmax+1],'r',alpha=0.5) # bottom
plt.plot(x[ymax,xmax:xmin-1:-1],y[ymax,xmax:xmin-1:-1],'r',alpha=0.5) # top
plt.plot(x[ymax:ymin-1:-1,xmin],y[ymax:ymin-1:-1,xmin],'r',alpha=0.5) # left
plt.plot(x[ymin:ymax+1,xmax],y[ymin:ymax+1,xmax],'r',alpha=0.5) # right
plt.xlabel("eje $x$")
plt.ylabel("eje $y$")
plt.axis('equal')
plt.show()

area=((xmax-xmin)*(ymax-ymin))

print("el area es {}".format(area))
print("el valor del rotor dentro del area {}".format(rotor))
print("circulacion ",circulacion)
print("rotor multiplicado por area ={}".format(rotor*area))

```

# Un modelo para los vientos huracanados y el campo magnetico
##  Ley Biot Savart
Definamos ahora un patrón de flujo de fluido diferente donde un campo rotacional disminuye en amplitud radialmente alejándose de la fuente. Este modelo es equivalente a los huracanes o al campo magnético alrededor de un cable vertical largo:
field around a long vertical wire:

```python
def B(I,x,y):                                         
    mu0 = 1.26 * 10**(-6) 
    r = np.sqrt((x)**2+(y)**2)
    c = mu0*I/(2*np.pi) 
    Bx = -y*c/r**2
    By = x*c/r**2         
    return Bx,By
```

cuando r vale cero explota.... y python se queja

```python
I=1
vx,vy = B(I,x,y)   
```

```python
xmin=3
xmax=17
ymin=3
ymax=17
```
