# Campo electrico en condiciones estaticas

Vamos a plotear en 2D el campo electrico existente entre 2 cargas q1 y q2 separadas del origen a una distancia L

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as npES

%matplotlib notebook
# set up a normalized grid:
dim= 20
xarray= np.arange(-dim,dim)
yarray= np.arange(-dim,dim)

#den = np.hypot(x-r0[0], y-r0[1])**3
q1=2
q2=-2
# (fluid) flow from a source at L to a sink at -L:
L = dim/2
x,y = np.meshgrid(xarray,yarray)
#vx = (x-L)/((x-L)**2+y**2) - (x+L)/((x+L)**2 +y**2)
vx= q1*(x-L)/(np.hypot(x-L,y)**3) + q2*(x+L)/(np.hypot(x+L,y)**3)
#vy = y/((x-L)**2+y**2) - y/((x+L)**2 +y**2)
vy = q1*y/(np.hypot(x-L,y)**3) + q2*y/(np.hypot(x+L,y)**3)
```

Ese error de Python es provocado porque el valor del campo electrico cerca de las cargas toma valores altos

```python
# plot the flow lines:
plt.figure()
#plt.quiver(x,y, vx, vy, pivot='mid')   # es un grafico un poco mas feo que el streamplot
plt.streamplot(x,y,vx,vy)
plt.xlabel("$x$-axis")
plt.ylabel("$y$-axis")
plt.axis('equal')
plt.show()
```

El flujo  $\Phi$ :

$$ \Phi_v = \iint {\bf E} \cdot d{\bf S}.$$

Definamos una caja calculemos su flujo. SegÃºn la ecuaciÃ³n anterior, este es el componente de flujo perpendicular a los lados de la caja. Para una caja alineada con los ejes x e y, el flujo a travÃ©s de los lados se define por Â± Ey, y la parte superior e inferior por Â± Ex:


```python
xmin=18
xmax=39
ymin=12
ymax=30

# total flux out the bottom and top:
topandbottomflux = -np.sum(vy[ymin,xmin:xmax+1])+np.sum(vy[ymax,xmin:xmax+1])

# total flux out the left and right side:
sideflux = -np.sum(vx[ymin:ymax+1,xmin])+np.sum(vx[ymin:ymax+1,xmax])

# net flux out of the box:
totalflux = sideflux + topandbottomflux

# plot flux and box:
plt.figure()
plt.title('El flujo tocal por la caja {:.3}'.format(totalflux))
# the flux:
#plt.quiver(x,y, vx, vy, pivot='mid')
plt.streamplot(x,y,vx,vy)
# bottom, top, left and right side of the box, respectively:
plt.plot(x[ymin,xmin:xmax+1],y[ymin,xmin:xmax+1],'r',alpha=0.5)
plt.plot(x[ymax,xmin:xmax+1],y[ymax,xmin:xmax+1],'r',alpha=0.5)
plt.plot(x[ymin:ymax+1,xmin],y[ymin:ymax+1,xmin],'r',alpha=0.5)
plt.plot(x[ymin:ymax+1,xmax],y[ymin:ymax+1,xmax],'r',alpha=0.5)
plt.xlabel("$x$-axis")
plt.ylabel("$y$-axis")
plt.axis('equal')
plt.show()
```

El flujo hacia el exterior por unidad de Ã¡rea (en 2D) o por unidad de volumen (en 3D) tambiÃ©n se denomina "la divergencia" del campo vectorial $ {\ bf E} $, y se puede escribir como $ \ nabla \ cdot {\ bf v} $. En coordenadas cartesianas 2D esto es

$$ \nabla\cdot {\bf v} = \frac{\partial v_x}{\partial x}+ \frac{\partial v_y}{\partial y}.$$

La divergencia para un Ã¡rea que no incluye un fuente o sumidero es cero. Puede confirmar esto, calculando la integral de superficie o podemos aproximarnos numÃ©ricamente:


```python
xmin=18
xmax=30
ymin=28
ymax=35

# total flux out the bottom and top:
topandbottomflux = -np.sum(vy[ymin,xmin:xmax+1])+np.sum(vy[ymax,xmin:xmax+1])

# total flux out the left and right side:
sideflux = -np.sum(vx[ymin:ymax+1,xmin])+np.sum(vx[ymin:ymax+1,xmax])

# net flux out of the box:
totalflux = sideflux + topandbottomflux

# plot flux and box:
plt.figure()
plt.title('El flujo tocal por la caja {:.3}'.format(totalflux))
# the flux:
#plt.quiver(x,y, vx, vy, pivot='mid')
plt.streamplot(x,y,vx,vy)
# bottom, top, left and right side of the box, respectively:
plt.plot(x[ymin,xmin:xmax+1],y[ymin,xmin:xmax+1],'r',alpha=0.5)
plt.plot(x[ymax,xmin:xmax+1],y[ymax,xmin:xmax+1],'r',alpha=0.5)
plt.plot(x[ymin:ymax+1,xmin],y[ymin:ymax+1,xmin],'r',alpha=0.5)
plt.plot(x[ymin:ymax+1,xmax],y[ymin:ymax+1,xmax],'r',alpha=0.5)
plt.xlabel("$x$-axis")
plt.ylabel("$y$-axis")
plt.axis('equal')
plt.show()
```

Para el caso del campo magnÃ©tico  ð, la divergencia es siempre cero (âˆ‡â‹…ð = 0),ya que nunca se han observado monopolos magnÃ©ticos.

We used the model of fluid flow as our vector field, but a distribution of charges would set up an electric ${\bf E}$ vector field, and masses would result in a gravitational field ${\bf g}$. For the case of the vector magnetic field ${\bf B}$, the divergence is always zero ($\nabla\cdot {\bf B}= 0$), because we either have not encompassed any sources/sinks, or we included a north and a south pole as magnetic monopoles have never been observed. 

# Divergencia de B

```python
def B(I,x,y):                                         
    mu0 = 1.26 * 10**(-6) 
    r = np.sqrt((x)**2+(y)**2)
    c = mu0*I/(2*np.pi) 
    Bx = -y*c/r**2
    By = x*c/r**2         
    return Bx,By
```

```python
I=10000000
Bx,By=B(I,x,y)
```

```python
fig = plt.figure()
plt.streamplot(x,y,Bx,By)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
